3.Calculate total revenue

```python
df=data['Returns']=data['Product Price']*data['Quantity']
```

```python
print(df)
```

4.Most Purchased Products

```python
top_products=data.groupby('Product Category')["Quantity"].sum().sort_values(ascending=False)
print(top_products)
```

5.Group customers based on spending.

```python
cus_spending=data.groupby('Customer ID')['Returns'].sum()
```

```python
print(cus_spending)
```

segments

```python
segments=pd.qcut(cus_spending,q=3,labels=['Low','Medium','High'])
```

```python

print(segments)
```

Average Order Value (AOV)

```python
total=data['Returns'].sum()
total_orders=data['Customer ID'].nunique()
AOV=total/total_orders
print("avg order value=",AOV)
```

Repeat Customer Analysis

```python
repeat_cus=data.groupby('Customer ID')['Order ID'].nunique()
repeat=repeat_cus[repeat_cus > 1]
print(repeat)
```
